```python
bar = (
    Bar()
    .add_xaxis(['Python', 'Java', '数据分析', '产品经理'])
    .add_yaxis("北京", ['25.1', '14.7', '24.6', '32.7'])
    .add_yaxis("上海", ['24.5', '32.4', '15.8', '23.3'])
    .add_yaxis("广州", ['14.5', '16.5', '14.6', '17.2'])
    .add_yaxis("深圳", ['25.3', '14.5', '32.6', '25.9'])
    .add_yaxis("成都", ['13.9', '15.9', '5.5', '16.1'])
    .add_yaxis("杭州", ['14.9', '32.8', '14.6', '24.9'])
    .add_yaxis("南京", ['14.4', '15.4', '9.2', '25'])
    .add_yaxis("武汉", ['12.4', '15.3', '5.6', '18'])
    .add_yaxis("西安", ['13.4', '15.7', '7.8', '15.6'])
    .set_global_opts(
        xaxis_opts = opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=-30)),
        title_opts = opts.TitleOpts(title="招聘普遍薪资分布",subtitle=""),
        yaxis_opts = opts.AxisOpts(
            axislabel_opts=opts.LabelOpts(formatter="{value}K")
        ),
    
    )
    .set_series_opts(
        label_opts = opts.LabelOpts(is_show=True)
    )
)
bar.render_notebook()
```

```python
geo = (
    Geo()
    .add_schema(maptype="china")
    .add("", [list(z) for z in zip(provinces, Python_values)])
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts(is_piecewise=True, max_=26, min_=12),
        title_opts=opts.TitleOpts(title="Python 需求情况"),
    )
)
geo.render_notebook()
```

```python
geo = (
    Geo()
    .add_schema(maptype="china")
    .add("",[list(z) for z in zip(provinces,total_values)],type_=ChartType.HEATMAP)
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(
        # is_piecewise 是否为分段
        visualmap_opts=opts.VisualMapOpts(is_piecewise=False, max_=25, min_=12),
        title_opts=opts.TitleOpts(title="需求热力图"),    
    )
)
geo.render_notebook()
```

```python
# beijing
beijing_python_jobname = beijing[beijing['job_type'] == 'python']['job_name'].value_counts()
beijing_python_jobname_dict = {'name': beijing_python_jobname.index, 'counts': beijing_python_jobname.values}
beijing_python_jobname_pd = pd.DataFrame(beijing_python_jobname_dict)

beijing_java_jobname = beijing[beijing['job_type'] == 'java']['job_name'].value_counts()
beijing_java_jobname_dict = {'name': beijing_java_jobname.index, 'counts': beijing_java_jobname.values}
beijing_java_jobname_pd = pd.DataFrame(beijing_java_jobname_dict)

beijing_dataanalyse_jobname = beijing[beijing['job_type'] == '数据分析']['job_name'].value_counts()
beijing_dataanalyse_jobname_dict = {'name': beijing_dataanalyse_jobname.index, 'counts': beijing_dataanalyse_jobname.values}
beijing_dataanalyse_jobname_pd = pd.DataFrame(beijing_dataanalyse_jobname_dict)

beijing_product_jobname = beijing[beijing['job_type'] == '产品经理']['job_name'].value_counts()
beijing_product_jobname_dict = {'name': beijing_product_jobname.index, 'counts': beijing_product_jobname.values}
beijing_product_jobname_pd = pd.DataFrame(beijing_product_jobname_dict)
```
